1. Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and
interpret the results

```python
import math
from scipy.stats import norm

def z_test(sample_mean, population_mean, population_std, sample_size, alpha=0.05):
    # Calculate the standard error
    standard_error = population_std / math.sqrt(sample_size)

    # Calculate the Z-score
    z_score = (sample_mean - population_mean) / standard_error

    # Calculate the p-value (two-tailed test)
    p_value = 2 * (1 - norm.cdf(abs(z_score)))

    # Interpretation
    print(f"Z-score: {z_score:.4f}")
    print(f"P-value: {p_value:.4f}")

    if p_value < alpha:
        print(f"Reject the null hypothesis at significance level {alpha}.")
        print("There is sufficient evidence to conclude the sample mean is significantly different from the population mean.")
    else:
        print(f"Fail to reject the null hypothesis at significance level {alpha}.")
        print("There is not sufficient evidence to conclude the sample mean is different from the population mean.")

# Example usage:
sample_mean = 102
population_mean = 100
population_std = 15
sample_size = 50
alpha = 0.05

z_test(sample_mean, population_mean, population_std, sample_size, alpha)
```

4. Perform a two-tailed Z-test using Python and visualize the decision region on a plot

```python
import math
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm

def two_tailed_z_test(sample_mean, population_mean, population_std, sample_size, alpha=0.05):
    # Calculate standard error
    standard_error = population_std / math.sqrt(sample_size)

    # Calculate Z-score
    z_score = (sample_mean - population_mean) / standard_error

    # Calculate two-tailed p-value
    p_value = 2 * (1 - norm.cdf(abs(z_score)))

    print(f"Z-score: {z_score:.4f}")
    print(f"P-value: {p_value:.4f}")

    # Critical z-values for two-tailed test
    z_critical_low = norm.ppf(alpha/2)
    z_critical_high = norm.ppf(1 - alpha/2)

    # Decision
    if z_score < z_critical_low or z_score > z_critical_high:
        decision = "Reject the null hypothesis"
    else:
        decision = "Fail to reject the null hypothesis"
    print(f"Decision: {decision} at significance level {alpha}")

    # Plot the normal distribution
    x = np.linspace(-4, 4, 1000)
    y = norm.pdf(x, 0, 1)

    plt.figure(figsize=(10,6))
    plt.plot(x, y, label='Standard Normal Distribution')

    # Shade rejection regions
    plt.fill_between(x, 0, y, where=(x <= z_critical_low), color='red', alpha=0.3, label='Rejection Region')
    plt.fill_between(x, 0, y, where=(x >= z_critical_high), color='red', alpha=0.3)

    # Plot the z-score
    plt.axvline(z_score, color='blue', linestyle='--', label=f'Z-score = {z_score:.2f}')

    # Add labels and legend
    plt.title('Two-Tailed Z-Test with Decision Regions')
    plt.xlabel('Z-value')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True)
    plt.show()

# Example usage:
sample_mean = 108
population_mean = 100
population_std = 15
sample_size = 50
alpha = 0.05

two_tailed_z_test(sample_mean, population_mean, population_std, sample_size, alpha)
```

5. Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
import math

def plot_type1_type2_errors(pop_mean, alt_mean, std_dev, sample_size, alpha=0.05):
    # Calculate standard error
    se = std_dev / math.sqrt(sample_size)

    # Critical value for rejecting H0 (two-tailed test)
    z_critical_low = norm.ppf(alpha/2)
    z_critical_high = norm.ppf(1 - alpha/2)

    # Corresponding critical values on the sample mean scale (null hypothesis)
    crit_val_low = pop_mean + z_critical_low * se
    crit_val_high = pop_mean + z_critical_high * se

    # Type 1 error (alpha) is given by alpha (red tails)
    # Type 2 error (beta) is probability that the test statistic falls inside critical values under alternative hypothesis

    # Calculate beta
    beta_low = norm.cdf(crit_val_high, loc=alt_mean, scale=se) - norm.cdf(crit_val_low, loc=alt_mean, scale=se)
    beta = beta_low

    # Power of test
    power = 1 - beta

    print(f"Type 1 error (Œ±): {alpha}")
    print(f"Type 2 error (Œ≤): {beta:.4f}")
    print(f"Power of the test (1 - Œ≤): {power:.4f}")

    # X range for plotting
    x = np.linspace(pop_mean - 4*se, alt_mean + 4*se, 1000)

    # Null hypothesis distribution
    null_dist = norm.pdf(x, loc=pop_mean, scale=se)
    # Alternative hypothesis distribution
    alt_dist = norm.pdf(x, loc=alt_mean, scale=se)

    plt.figure(figsize=(12,6))

    # Plot null distribution
    plt.plot(x, null_dist, label='Null Hypothesis (H0)', color='blue')
    # Plot alternative distribution
    plt.plot(x, alt_dist, label='Alternative Hypothesis (H1)', color='green')

    # Shade Type 1 error regions (tails under null)
    plt.fill_between(x, 0, null_dist, where=(x <= crit_val_low), color='red', alpha=0.3, label='Type 1 Error (Œ±)')
    plt.fill_between(x, 0, null_dist, where=(x >= crit_val_high), color='red', alpha=0.3)

    # Shade Type 2 error region (area under alternative inside critical region)
    plt.fill_between(x, 0, alt_dist, where=((x > crit_val_low) & (x < crit_val_high)), color='orange', alpha=0.3, label='Type 2 Error (Œ≤)')

    plt.axvline(crit_val_low, color='red', linestyle='--', label='Critical Values')
```

8.  Simulate data and perform both Z-test and T-test, then compare the results using Python

```python
import numpy as np
import math
from scipy.stats import norm, t

# Simulate sample data
np.random.seed(42)
population_mean = 100
population_std = 15  # known population std dev

sample_size = 40
# Generate sample data with true mean slightly different from population mean
sample_data = np.random.normal(loc=103, scale=population_std, size=sample_size)

# Calculate sample statistics
sample_mean = np.mean(sample_data)
sample_std = np.std(sample_data, ddof=1)  # sample std dev

alpha = 0.05

# One-sample Z-test (population std dev known)
def one_sample_z_test(sample_mean, population_mean, population_std, sample_size, alpha=0.05):
    se = population_std / math.sqrt(sample_size)
    z_score = (sample_mean - population_mean) / se
    p_value = 2 * (1 - norm.cdf(abs(z_score)))
    return z_score, p_value

# One-sample T-test (population std dev unknown)
def one_sample_t_test(sample_data, population_mean, alpha=0.05):
    n = len(sample_data)
    sample_mean = np.mean(sample_data)
    sample_std = np.std(sample_data, ddof=1)
    se = sample_std / math.sqrt(n)
    t_score = (sample_mean - population_mean) / se
    p_value = 2 * (1 - t.cdf(abs(t_score), df=n-1))
    return t_score, p_value

# Run tests
z_score, p_z = one_sample_z_test(sample_mean, population_mean, population_std, sample_size, alpha)
t_score, p_t = one_sample_t_test(sample_data, population_mean, alpha)

print(f"Sample mean: {sample_mean:.3f}")
print("--- One-sample Z-test ---")
print(f"Z-score: {z_score:.4f}")
print(f"P-value: {p_z:.4f}")
print("Reject null hypothesis" if p_z < alpha else "Fail to reject null hypothesis")

print("\n--- One-sample T-test ---")
print(f"T-score: {t_score:.4f}")
print(f"P-value: {p_t:.4f}")
print("Reject null hypothesis" if p_t < alpha else "Fail to reject null hypothesis")
```

9.  Write a Python function to calculate the confidence interval for a sample mean and explain its significance.

```python
import numpy as np
from scipy.stats import t

def confidence_interval(sample, confidence=0.95):
    """
    Calculate the confidence interval for the sample mean.

    Parameters:
        sample (array-like): The sample data.
        confidence (float): Confidence level (default 0.95 for 95% CI).

    Returns:
        (lower_bound, upper_bound): Tuple representing the confidence interval.
    """
    n = len(sample)
    mean = np.mean(sample)
    std_err = np.std(sample, ddof=1) / np.sqrt(n)  # standard error

    # Degrees of freedom
    df = n - 1

    # t critical value for two-tailed test
    t_crit = t.ppf((1 + confidence) / 2, df)

    margin_of_error = t_crit * std_err

    lower_bound = mean - margin_of_error
    upper_bound = mean + margin_of_error

    return lower_bound, upper_bound

# Example usage:
import random
random.seed(42)
sample_data = [random.gauss(100, 15) for _ in range(30)]

ci_low, ci_high = confidence_interval(sample_data, confidence=0.95)
print(f"95% Confidence Interval: ({ci_low:.2f}, {ci_high:.2f})")
```

10. Write a Python program to calculate the margin of error for a given confidence level using sample data

```python
import numpy as np
from scipy.stats import t

def margin_of_error(sample, confidence=0.95):
    """
    Calculate the margin of error for the sample mean at the specified confidence level.

    Parameters:
        sample (array-like): The sample data.
        confidence (float): Confidence level (default 0.95 for 95% confidence).

    Returns:
        float: The margin of error.
    """
    n = len(sample)
    std_err = np.std(sample, ddof=1) / np.sqrt(n)  # standard error
    df = n - 1
    t_crit = t.ppf((1 + confidence) / 2, df)  # t critical value

    margin = t_crit * std_err
    return margin

# Example usage
import random
random.seed(42)
sample_data = [random.gauss(50, 10) for _ in range(30)]

moe = margin_of_error(sample_data, confidence=0.95)
print(f"Margin of Error at 95% confidence: {moe:.3f}")
```

11.  Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process


```python
def bayes_theorem(prior_H, likelihood_E_given_H, prior_not_H, likelihood_E_given_not_H):
    """
    Calculate the posterior probability P(H|E) using Bayes' theorem.

    Parameters:
    - prior_H: P(H), prior probability of hypothesis H
    - likelihood_E_given_H: P(E|H), likelihood of evidence given H
    - prior_not_H: P(~H), prior probability of not H
    - likelihood_E_given_not_H: P(E|~H), likelihood of evidence given not H

    Returns:
    - posterior probability P(H|E)
    """
    numerator = likelihood_E_given_H * prior_H
    denominator = numerator + likelihood_E_given_not_H * prior_not_H
    posterior = numerator / denominator
    return posterior

# Given data:
prior_disease = 0.01           # P(Disease)
prior_no_disease = 1 - prior_disease  # P(No Disease)

sensitivity = 0.99             # P(Positive test | Disease)
false_positive_rate = 1 - 0.95 # P(Positive test | No Disease)

# Calculate posterior probability of disease given positive test
posterior = bayes_theorem(prior_disease, sensitivity, prior_no_disease, false_positive_rate)

print(f"Probability of having disease given a positive test result: {posterior:.4f}")
```

```python
import numpy as np
import pandas as pd

def calculate_expected_frequencies(observed):
    """
    Calculate expected frequencies for a Chi-square test.

    Parameters:
        observed (2D array or DataFrame): Observed frequency table.

    Returns:
        expected (DataFrame): Expected frequency table.
    """
    observed = np.array(observed)

    # Row and column totals
    row_totals = observed.sum(axis=1).reshape(-1, 1)
    col_totals = observed.sum(axis=0).reshape(1, -1)
    grand_total = observed.sum()

    # Expected frequencies formula: E_ij = (row_total_i * col_total_j) / grand_total
    expected = (row_totals @ col_totals) / grand_total
    return expected

# Example observed data: (2 categories x 3 preferences)
observed_data = np.array([[20, 30, 25],  # Male
                          [25, 35, 15]]) # Female

# Calculate expected frequencies
expected = calculate_expected_frequencies(observed_data)

# Display using DataFrame for clarity
df_expected = pd.DataFrame(expected, index=['Male', 'Female'], columns=['Like', 'Neutral', 'Dislike'])
print("Expected Frequencies:\n")
print(df_expected.round(2))
```

16. Implement an F-test using Python to compare the variances of two random samples

```python
import numpy as np
from scipy.stats import f

def f_test(sample1, sample2, alpha=0.05):
    # Sample sizes
    n1, n2 = len(sample1), len(sample2)

    # Sample variances (unbiased)
    var1, var2 = np.var(sample1, ddof=1), np.var(sample2, ddof=1)

    # Determine which variance is larger for F = larger/smaller
    if var1 > var2:
        f_stat = var1 / var2
        df1, df2 = n1 - 1, n2 - 1
    else:
        f_stat = var2 / var1
        df1, df2 = n2 - 1, n1 - 1

    # One-tailed test for variance inequality
    p_value = 2 * (1 - f.cdf(f_stat, df1, df2))

    print(f"Sample Variance 1: {var1:.4f}")
    print(f"Sample Variance 2: {var2:.4f}")
    print(f"F-statistic: {f_stat:.4f}")
    print(f"P-value: {p_value:.4f}")

    if p_value < alpha:
        print(f"\nReject the null hypothesis at Œ± = {alpha}")
        print("Variances are significantly different.")
    else:
        print(f"\nFail to reject the null hypothesis at Œ± = {alpha}")
        print("No significant difference between the variances.")

# Example: Simulate two samples
np.random.seed(42)
sample1 = np.random.normal(loc=50, scale=10, size=30)
sample2 = np.random.normal(loc=50, scale=15, size=30)

f_test(sample1, sample2)
```

19. Write a Python function to check the assumptions (normality, independence, and equal variance) for ANOVAD

```python
import numpy as np
import pandas as pd
from scipy.stats import shapiro, levene
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
from statsmodels.formula.api import ols

def check_anova_assumptions(data, group_col, value_col):
    """
    Checks ANOVA assumptions:
      - Normality of residuals
      - Homogeneity of variances
      - Independence (assumed by design)

    Parameters:
        data (DataFrame): pandas DataFrame containing the data
        group_col (str): name of the column containing group labels
        value_col (str): name of the column containing numerical values
    """

    # Fit ANOVA model
    model = ols(f'{value_col} ~ C({group_col})', data=data).fit()
    residuals = model.resid

    print("1. Checking Normality of Residuals (Shapiro-Wilk Test)")
    stat, p = shapiro(residuals)
    print(f"   Shapiro-Wilk Test p-value: {p:.4f}")
    if p > 0.05:
        print("   ‚úî Residuals appear to be normally distributed.\n")
    else:
        print("   ‚úò Residuals may not be normally distributed.\n")

    print("2. Checking Homogeneity of Variance (Levene‚Äôs Test)")
    grouped_data = [group[value_col].values for name, group in data.groupby(group_col)]
    stat, p = levene(*grouped_data)
    print(f"   Levene‚Äôs Test p-value: {p:.4f}")
    if p > 0.05:
        print("   ‚úî Variances are approximately equal across groups.\n")
    else:
        print("   ‚úò Variances are not equal across groups.\n")

    print("3. Independence of Observations")
    print("   ‚úî Assumed by design (must be ensured during data collection).\n")

    # Optional: Residual Q-Q plot
    sm.qqplot(residuals, line='s')
    plt.title('Q-Q Plot of ANOVA Residuals')
    plt.grid(True)
    plt.show()

    # Optional: Residuals vs. Fitted plot
    fitted_vals = model.fittedvalues
    plt.figure(figsize=(6, 4))
    sns.residplot(x=fitted_vals, y=residuals, lowess=True, line_kws={'color': 'red'})
    plt.xlabel("Fitted values")
    plt.ylabel("Residuals")
    plt.title("Residuals vs Fitted")
    plt.grid(True)
    plt.show()
```

21. Write a Python program to visualize the F-distribution and discuss its use in hypothesis testing

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import f

def plot_f_distribution(df1, df2, alpha=0.05):
    """
    Visualize the F-distribution and mark the critical value for given degrees of freedom and alpha.

    Parameters:
    df1 (int): Degrees of freedom for the numerator
    df2 (int): Degrees of freedom for the denominator
    alpha (float): Significance level (default = 0.05)
    """
    # Generate F-distribution values
    x = np.linspace(0, 5, 1000)
    y = f.pdf(x, df1, df2)

    # Critical value for one-tailed test (right tail)
    critical_value = f.ppf(1 - alpha, df1, df2)

    # Plot the distribution
    plt.figure(figsize=(10, 6))
    plt.plot(x, y, label=f'F-distribution (df1={df1}, df2={df2})', color='blue')
    plt.fill_between(x, y, where=(x >= critical_value), color='red', alpha=0.5,
                     label=f'Œ± = {alpha} (Critical region)')

    plt.axvline(critical_value, color='red', linestyle='--', label=f'Critical Value = {critical_value:.2f}')
    plt.title('F-Distribution and Critical Region')
    plt.xlabel('F-value')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True)
    plt.show()

# Example usage
plot_f_distribution(df1=4, df2=30, alpha=0.05)
```

24.  Perform a hypothesis test for population variance using a Chi-square distribution and interpret the resultsD

```python
import numpy as np
from scipy.stats import chi2

# Sample data (simulate or use your own)
np.random.seed(42)
sample = np.random.normal(loc=100, scale=15, size=30)  # Mean=100, SD=15

# Given hypothesized population variance
sigma0_squared = 15**2  # Hypothesized variance = 225

# Sample statistics
n = len(sample)
sample_var = np.var(sample, ddof=1)  # Sample variance (unbiased)

# Test statistic
chi_stat = (n - 1) * sample_var / sigma0_squared

# Degrees of freedom
df = n - 1

# Significance level
alpha = 0.05

# Critical values for two-tailed test
chi_lower = chi2.ppf(alpha/2, df)
chi_upper = chi2.ppf(1 - alpha/2, df)

# p-value calculation (two-tailed)
p_value = 2 * min(chi2.cdf(chi_stat, df), 1 - chi2.cdf(chi_stat, df))

print("Chi-square test for population variance")
print("---------------------------------------")
print(f"Sample variance: {sample_var:.2f}")
print(f"Test statistic (Chi-square): {chi_stat:.2f}")
print(f"Degrees of freedom: {df}")
print(f"Critical values: {chi_lower:.2f} (lower), {chi_upper:.2f} (upper)")
print(f"P-value: {p_value:.4f}")

if chi_stat < chi_lower or chi_stat > chi_upper:
    print("\nReject the null hypothesis.")
    print("There is evidence that the population variance differs from the hypothesized value.")
else:
    print("\nFail to reject the null hypothesis.")
    print("No evidence to suggest the population variance differs from the hypothesized value.")
```

25. Write a Python script to perform a Z-test for comparing proportions between two datasets or groups

```python
import numpy as np
from statsmodels.stats.proportion import proportions_ztest

def two_proportion_ztest(count1, nobs1, count2, nobs2, alpha=0.05):
    """
    Perform a two-proportion Z-test.

    Parameters:
    - count1, count2: number of successes in group 1 and 2
    - nobs1, nobs2: total observations in group 1 and 2
    - alpha: significance level (default 0.05)

    Returns:
    - z_stat: Z test statistic
    - p_value: two-tailed p-value
    """
    counts = np.array([count1, count2])
    nobs = np.array([nobs1, nobs2])

    z_stat, p_value = proportions_ztest(counts, nobs)

    print("Two-Proportion Z-Test")
    print("---------------------")
    print(f"Group 1: {count1} successes out of {nobs1} observations")
    print(f"Group 2: {count2} successes out of {nobs2} observations")
    print(f"Z-statistic: {z_stat:.4f}")
    print(f"P-value: {p_value:.4f}")

    if p_value < alpha:
        print(f"\nReject the null hypothesis at alpha={alpha}.")
        print("There is a significant difference between the two proportions.")
    else:
        print(f"\nFail to reject the null hypothesis at alpha={alpha}.")
        print("No significant difference between the two proportions.")

    return z_stat, p_value

# Example usage
two_proportion_ztest(count1=45, nobs1=100, count2=30, nobs2=100)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import f

# Simulate two datasets
np.random.seed(123)
data1 = np.random.normal(loc=50, scale=10, size=40)  # SD = 10
data2 = np.random.normal(loc=50, scale=15, size=35)  # SD = 15

# Sample variances and sizes
var1 = np.var(data1, ddof=1)
var2 = np.var(data2, ddof=1)
n1 = len(data1)
n2 = len(data2)

# Arrange so that var1 >= var2 for F >= 1
if var1 >= var2:
    F_stat = var1 / var2
    dfn, dfd = n1 - 1, n2 - 1
else:
    F_stat = var2 / var1
    dfn, dfd = n2 - 1, n1 - 1

# Significance level
alpha = 0.05

# Critical values for two-tailed test
F_lower = f.ppf(alpha / 2, dfn, dfd)
F_upper = f.ppf(1 - alpha / 2, dfn, dfd)

# p-value calculation (two-tailed)
p_value = 2 * min(f.cdf(F_stat, dfn, dfd), 1 - f.cdf(F_stat, dfn, dfd))

# Print results
print("F-test for comparing two variances")
print("----------------------------------")
print(f"Variance 1: {var1:.2f}, Sample size 1: {n1}")
print(f"Variance 2: {var2:.2f}, Sample size 2: {n2}")
print(f"F-statistic: {F_stat:.4f}")
print(f"Degrees of freedom: ({dfn}, {dfd})")
print(f"Critical values: Lower = {F_lower:.4f}, Upper = {F_upper:.4f}")
print(f"P-value: {p_value:.4f}")

if F_stat < F_lower or F_stat > F_upper:
    print("\nReject the null hypothesis: variances are significantly different.")
else:
    print("\nFail to reject the null hypothesis: no significant difference in variances.")

# Visualization
x = np.linspace(0, 5, 1000)
y = f.pdf(x, dfn, dfd)

plt.figure(figsize=(10, 6))
plt.plot(x, y, label=f'F-distribution df=({dfn},{dfd})', color='blue')
plt.fill_between(x, y, where=(x <= F_lower), color='red', alpha=0.5, label='Lower rejection region')
plt.fill_between(x, y, where=(x >= F_upper), color='red', alpha=0.5, label='Upper rejection region')
plt.axvline(F_stat, color='green', linestyle='--', label=f'F statistic = {F_stat:.2f}')
plt.title('F-distribution with Rejection Regions and Test Statistic')
plt.xlabel('F value')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()
```
